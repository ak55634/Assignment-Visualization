```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

#  uploaded file 'CAREGIVERS.csv'
caregivers = pd.read_csv('CAREGIVERS.csv')

# Display the first few rows
print(caregivers.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count of Caregiver Descriptions
sns.countplot(data=caregivers, y='DESCRIPTION', order=caregivers['DESCRIPTION'].value_counts().index)
plt.title('Caregiver Descriptions')
plt.xlabel('Count')
plt.ylabel('Description')
plt.show()
```
