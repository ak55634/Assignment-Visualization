```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

# uploaded files 'ADMISSIONS.csv' and 'PATIENTS.csv'
admissions_df = pd.read_csv('ADMISSIONS.csv')
patients_df = pd.read_csv('PATIENTS.csv')

# Display the first few rows
#print(admissions.head())
#print(patients.head())


```

```python
#  date columns to datetime
patients_df['DOB'] = pd.to_datetime(patients_df['DOB'])
admissions_df['ADMITTIME'] = pd.to_datetime(admissions_df['ADMITTIME'])

# Merge the data on the common patient
combined_df = pd.merge(patients_df, admissions_df, on='SUBJECT_ID')

# Set 'ADMITTIME'
combined_df.set_index('ADMITTIME', inplace=True)

plt.figure(figsize=(12, 6))

# Plot the number of admissions over time
combined_df.resample('ME').size().plot(label='Number of Admissions', color='blue')

plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.title('Time Series Plot of Admissions Over Time')
plt.legend()
plt.show()
```
