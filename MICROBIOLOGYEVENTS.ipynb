```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

# Assuming the uploaded file is named 'MICROBIOLOGYEVENTS.csv'
microbiology = pd.read_csv('MICROBIOLOGYEVENTS.csv')

# Display the first few rows of the DataFrame to inspect the data
print(microbiology.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Chart of Top 10 Microbiology Test Types
top_10_tests = microbiology['SPEC_TYPE_DESC'].value_counts().head(5)
sns.barplot(x=top_10_tests.values, y=top_10_tests.index)
plt.title('Top 5 Test Types')
plt.xlabel('Count')
plt.ylabel('Test Type')
plt.show()
```
